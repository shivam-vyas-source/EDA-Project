**Dividing data between categorical value and numerical value**


```python
from pandas.core.interchange.dataframe_protocol import Column
categorical=[]
numerical=[]
for column in Data.columns:
  if Data[column].nunique() > 10:
    numerical.append(Column)
  else:
    categorical.append(column)
```

**Univariate Analysis**

```python
# find aggregate sales among Branches...
plt.hist(Data['Branch'])
```

```python
Data['Branch'].value_counts().plot(kind="pie",autopct="%1.2f%%")
```

**Bivariate Analysis / Multivariate Analysis**

```python
# Does the cost of goods sold Affect the ratings that the customers provide
sns.scatterplot(x=Data['cogs'],y=Data['Rating'])
```

```python
# Does gross income affect the Ratings that the customers provide
sns.scatterplot(x=Data['gross income'],y=Data['Rating'])
```

```python
# find the most profitable Branch As per Gross Income
sns.barplot(x=Data['gross income'],y=Data['Branch'],hue=Data['City'])
```

```python
# is there any relationship between gender and gross income
sns.boxplot(x=Data['Gender'],y=Data['gross income'],hue=Data['Customer type'])
```

```python
# Find the product line that generates the most income...
sns.barplot(x=Data['Product line'],y=Data['gross income'])
plt.xticks(rotation=75)
plt.yticks(rotation=70)
plt.show()
```

```python
# Find the highest unit price in the product line...
sns.barplot(x=Data['Product line'],y=Data['Unit price'])
plt.xticks(rotation=75)
plt.yticks(rotation=70)
plt.show()
```

```python
dw_mapping={
    0:'mon',1:'tue',2:'wed',3:'thur',4:'fri',5:'sat',6:'sun'}
```

```python
# Display daily sales by day of week..
Data['day_of_week']=Data['Date'].dt.dayofweek.map(dw_mapping)
Data['day_of_week'].value_counts().plot(kind='bar')
```

```python
# what will be the highest sales for the months
month_mapping={
    1:'jan',2:'tue',3:'wed',4:'thu',5:'fri',6:'sat'}
```

```python
Data['month']=Data['Date'].dt.month.map(month_mapping)
```

```python
Data['month'].value_counts().plot(kind='bar')
```
